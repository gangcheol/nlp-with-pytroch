* 예를 들어

    "모스크바에 여행을 왔는데 건물도 예쁘고 먹을 것도 맛있었어. 그런데 글쎄 직장 상사한테 전화가 왔어. 어디냐고 묻더라구 그래서 
    나는 말했지. 저 여행왔는데요. 여기 ___"

* 위와 같은 빈칸에는 **"모스크바"** 라는 단어가 와야 적당하지만 RNN이 충분한 기억력을 가지고 있지 못한다면 엉뚱한 단어를 예측하게 된다.

* 위 같은 문제를 **"vanishing gradient problem","장기 의존성 문제점(the problem of Long-Term Dependencies)"** 이라고 한다.

# Long Short-Term Memory, LSTM

* input, forget, output gate 로 구성

* 삭제 게이트에서 이전 시점의 입력을 얼마나 반영할지를 결정할 수 있는 게이트를 추가해 RNN에서 발생하는 문제점을 해결한다.

# Cated Recurrent Units, GRU

* LSTM이 출력, 입력, 삭제 게이트가 존재했다면

* GRU에서는 업데이트 게이트와, 리셋 게이트 두가지가 존재한다.

* 경험적으로? LSTM과  GRU는 비슷한 성능을 낸다

# Attention Mechanism

* 디코더에서 출력단어를 예측하는 매 시점마다, 전체 입력 문장이 아닌, 해당 시점에서 예측해야할 단어와 연관성이 있는 입력 단어 부분을 좀 더 집중(Attention)해서 보는 것이다.

* 즉, 전체가 아닌 연관성이 있는 일부를 봄으로써 RNN의 고질적 문제인 기울기 소실 문제를 해결할 수 있다.

# Reinforcement Learning for Sequence Generation

* Teacher forcing : 기존의 RNN에서는 현시점에 값을 예측할 때 이전 단계에 정답 단어들을 입력값으로 넣어 학습시킨다.

* but, 위와 같은 경우 추론단계에서 이전 단계에 단어들이 정답 단어들이 아닌 모델자체에서 생성된 추론 값으로 대체된다.

* 즉, 생성된 모델을 가지고 test를 할경우 실제값이 아닌 모델에서 생성된 추론값을 사용하기 때문에 "exposure bias"이 발생한다.
